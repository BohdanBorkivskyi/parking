```python
#import cv2
from keras.preprocessing.image import load_img 
import numpy as np
img = load_img('test/2018-07-18 02_24_29.627+88.jpg')
pred = model.predict_classes(np.expand_dims(img, axis=0) / 255)
print(pred)
```

```python
import glob
import matplotlib.pyplot as plt


for i, filename in enumerate(glob.glob('parklots/validation/occupied/*')):
    img = load_img(filename)
    prediction = model.predict_classes(np.expand_dims(img, axis=0) / 255)[0][0]
    if prediction == 0:
        print(filename)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(img)
        plt.show()

```

```python
for i, filename in enumerate(glob.glob('parklots/validation/empty/*')):
    img = load_img(filename)
    prediction = model.predict_classes(np.expand_dims(img, axis=0) / 255)[0][0]
    if prediction == 1:
        print(filename)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(img)
        plt.show()
```
